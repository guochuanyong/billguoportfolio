```python
import pandas as pd

# === SETTINGS ===
CSV_FILE = "Community_District_Boundaries_20260223.csv"

# Put the row numbers you want to KEEP here (Excel-style row numbers)
# Example: keep row 1, 5, 6, 10
rows_to_keep = [1, 193, 221, 270, 314]

# =================

# Read CSV
df = pd.read_csv(CSV_FILE)

# Convert Excel-style row numbers to DataFrame indices
# Excel row 1 = header
# Excel row 2 = df index 0
# So: index = row_number - 2
indices_to_keep = [r - 2 for r in rows_to_keep if r > 1]

# Filter dataframe
df_filtered = df.iloc[indices_to_keep]

# Overwrite same file
df_filtered.to_csv(CSV_FILE, index=False)

print("Done. File overwritten with selected rows only.")
```

```python
import pandas as pd

CSV_FILE = "communities_with_geojson.csv"

# Read file WITHOUT treating first row as header
df = pd.read_csv(CSV_FILE, header=None)

# Drop the first row (row index 0)
df = df.iloc[1:]

# Overwrite same file WITHOUT header
df.to_csv(CSV_FILE, index=False, header=False)

print("Header row removed. File overwritten.")
```
